***

**Objetivos:** Explorar dados de vendas mensais de shampoo durante um período de 3 anos.

- Crie uma visualização que mostre a série temporal das vendas ao longo dos meses.
Existe alguma tendência ou alguma periodicidade?

```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

```python
vendas = pd.read_csv('dadosoficina07.txt')
vendas.head()
```

```python
vendas.describe()
```

- Gerando gráficos para análise

```python
plt.figure(figsize=(18,5))
plt.title('Vendas mensais de shampoo')
sns.scatterplot(x='Month', y='Sales', data=vendas)
plt.show()
```

- Observa-se acréscimo na vendas a partir de outrubro, porém sem a linha de continuidade dificulta ver essa evolução.

```python
plt.figure(figsize=(18,5))
plt.title('Venas mensais de shampoo')
sns.lineplot(x='Month',y='Sales', data=vendas)
plt.show()
```

- Com o visual de linha sequencial torna mais visível que no decorrer dos anos as vendas veem aumentando e com uma tendência de sempre haver picos de vendas entre meses de setembro e novembro.
